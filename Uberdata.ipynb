```python
# VISUALISATION 2
df['Status'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Trip Status Counts")
plt.ylabel("Count")
plt.xlabel("Status")
plt.show()
```

```python
# VISUALISATION 3
df['Pickup_Point'].value_counts().plot(kind='bar', color='orange')
plt.title("Pickup Point Distribution")
plt.ylabel("Trips")
plt.xlabel("Pickup Location")
plt.show()
```

```python
# VISUALISATION 4
df['Request_Timestamp'] = pd.to_datetime(df['Request_Timestamp'])
df['Drop_Timestamp'] = pd.to_datetime(df['Drop_Timestamp'])

df['hour'] = df['Request_Timestamp'].dt.hour
df['day'] = df['Request_Timestamp'].dt.day_name()
```

```python
df['day'].value_counts().sort_index().plot(kind='bar', figsize=(10,5))
plt.title("Trip Requests by Day")
plt.xlabel("Day")
plt.ylabel("Requests")
plt.show()
```

```python
# VISUALISATION 5
sns.countplot(data=df, x='hour', hue='Status', palette='Set2')
plt.title("Trip Status by Hour")
plt.show()
```

```python
# VISUALISATION 6
df['Status'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, figsize=(6,6))
plt.title("Trip Status Distribution")
plt.ylabel("")
plt.show()
```

```python
# VISUALISATION 7
cancel_pct = df[df['Status'] == 'Cancelled']['Pickup_Point'].value_counts(normalize=True) * 100
cancel_pct.plot(kind='bar', color='crimson')
plt.title("Cancellation % by Pickup Point")
plt.ylabel("Percentage")
plt.show()
```

```python
# VISUALISATION 8
df['Hour'] = df['Request_Timestamp'].dt.hour
gap = df.groupby(['hour', 'Pickup_Point'])['Status'].value_counts().unstack().fillna(0)
gap['Gap'] = gap.sum(axis=1) - gap.get('Trip Completed', 0)
gap['Gap'].unstack().plot(kind='bar', stacked=True, figsize=(12,6))
plt.title("Hourly Demand Gap by Pickup Point")
plt.ylabel("Gap (Requests - Completed)")
plt.show()
```

```python
# VISUALISATION 9
df_sorted = df.sort_values('Request_Timestamp')
df_sorted['Cumulative Trips'] = range(1, len(df_sorted)+1)
plt.plot(df_sorted['Request_Timestamp'], df_sorted['Cumulative Trips'], color='teal')
plt.title("Cumulative Trip Requests Over Time")
plt.xlabel("Time")
plt.ylabel("Total Trips")
plt.show()
```

```python
# VISUALISATION 10
cancel_data = df[df['Status'] == 'Cancelled'].groupby(['Hour', 'Pickup_Point']).size().unstack().fillna(0)
sns.heatmap(cancel_data, annot=True, fmt='.0f', cmap='Reds')
plt.title("Cancelled Trips by Hour & Pickup Point")
plt.show()
```

```python
# VISUALISATION 11
df['day'] = df['Request_Timestamp'].dt.day_name()
comp_rate = df[df['Status'] == 'Trip Completed'].groupby('day').size() / df.groupby('day').size() * 100
comp_rate = comp_rate[['Monday','Tuesday','Wednesday','Thursday','Friday']]
comp_rate.plot(kind='bar', color='green')
plt.title("Trip Completion Rate by Day")
plt.ylabel("Completion %")
plt.show()
```

```python
# VISUALISATION 12
df['Drop_Timestamp'] = pd.to_datetime(df['Drop_Timestamp'])
df['duration'] = (df['Drop_Timestamp'] - df['Request_Timestamp']).dt.total_seconds() / 60
avg_duration = df.groupby('Hour')['duration'].mean()
avg_duration.plot(kind='line', marker='o', color='navy')
plt.title("Average Trip Duration by Hour")
plt.ylabel("Duration (minutes)")
plt.grid(True)
plt.show()
```

```python
# VISUALISATION 13
no_car_ratio = df[df['Status'] == 'No Cars Available'].groupby('Hour').size() / df.groupby('Hour').size() * 100
no_car_ratio.plot(kind='bar', color='orange')
plt.title("No Cars Available % by Hour")
plt.ylabel("Percentage")
plt.show()
```

```python
# VISUALISATION 14
df['is_weekend'] = df['day'].isin(['Saturday', 'Sunday'])
sns.countplot(data=df, x='Hour', hue='is_weekend', palette='Set2')
plt.title("Hourly Trips: Weekday vs Weekend")
plt.show()
```

```python
# VISUALISATION 15
hourly_demand = df.groupby('Request_Timestamp').size().resample('H').sum()
hourly_demand.rolling(window=3).mean().plot(figsize=(12,4), color='dodgerblue')
plt.title("Hourly Demand Trend with Moving Average")
plt.ylabel("Requests")
plt.show()
```

```python
# VISUALISATION 16
pickup_day = df.groupby(['day', 'Pickup_Point']).size().unstack().fillna(0)
pickup_day = pickup_day.loc[['Monday','Tuesday','Wednesday','Thursday','Friday']]
sns.heatmap(pickup_day, annot=True, fmt='.0f', cmap='BuGn')
plt.title("Pickup Point Usage by Day")
plt.show()
```

```python
# VISUALISATION 17
sns.boxplot(data=df, x='Status', y='duration', palette='muted')
plt.title("Trip Duration by Status")
plt.ylim(0, 60)  # Optional to remove outliers
plt.show()
```

```python
# VISUALISATION 18
comp_ratio = df[df['Status']=='Trip Completed'].groupby('Hour').size() / df.groupby('Hour').size() * 100
comp_ratio.plot(marker='o', color='limegreen')
plt.title("Trip Completion % by Hour")
plt.ylabel("Completion %")
plt.show()
```

```python
# VISUALISATION 19
lost_trips = df[df['Status'] == 'No Cars Available']
estimated_loss = lost_trips.groupby('Hour').size() * 100
estimated_loss.plot(kind='bar', color='firebrick')
plt.title("Estimated Revenue Loss by Hour ()")
plt.ylabel(" Lost")
plt.show()
```

```python
# VISUALISATION 20
pickup_grp = df.groupby('Pickup_Point')['Status'].value_counts(normalize=True).unstack().fillna(0)
pickup_grp[['Cancelled', 'Trip Completed']].plot(kind='bar', stacked=True)
plt.title("Cancelled vs Completed Ratio by Pickup Point")
plt.ylabel("Ratio")
plt.show()
```

```python
# VISUALISATION 21
status_hourly = df.groupby(['hour', 'Status']).size().unstack().fillna(0)
status_hourly.plot(figsize=(12,5), marker='o')
plt.title("Hourly Breakdown of Each Trip Status")
plt.ylabel("Trips")
plt.grid(True)
plt.show()
```
